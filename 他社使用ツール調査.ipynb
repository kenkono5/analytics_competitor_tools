```
import requests
from bs4 import BeautifulSoup as bs
from time import sleep

from google.colab import files
uploaded = files.upload()

import pandas as pd

# df = pandas.read_excel('test.xlsx', sheet_name=0)
# df.to_csv('test.csv', index=False)

df= pd.read_csv('./①FUMA_東京_小売全般_マスターデータ.csv', encoding="utf_8")
df.head()

array = df.values
company_name_list = []
url_list = []

pardot_list = []
eloqua_list = []
marketo_list = []
urlNG_list = []
num = 1
for value in array:
  company_name_list.append(value[3])
  url_list.append(value[4])

for i, j in zip(url_list, company_name_list):
  try:
    res = requests.get(i, timeout=10)
    print(num)
  except Exception as e: 
    pass
    print("=== エラー内容 始 ===")
    print("type:" + str(type(e)))
    print("args:" + str(e.args))
    print("e自身:" + str(e))
    print("=== エラー内容 完 ===")
    urlNG_list.append(j)
    print(num, "pass!") 
  soup = bs(res.text, 'html.parser')
  script_tag = soup.find_all('script')
  num += 1
  for script in script_tag:
    if "pardot" in script.text:
      pardot_list.append(j)

    if "elqQ" in script.text:
      eloqua_list.append(j)

    if "marketo" in script.text:
      marketo_list.append(j)

print(pardot_list)
print(eloqua_list)
print(marketo_list)
print(urlNG_list)

use_pardot = []
use_eloqua = []
use_marketo = []
ng_url = []
for value in array:
  if value[3] in pardot_list:
    use_pardot.append(1)
  else:
    use_pardot.append(0)

  if value[3] in eloqua_list:
    use_eloqua.append(1)
  else:
    use_eloqua.append(0)

  if value[3] in marketo_list:
    use_marketo.append(1)
  else:
    use_marketo.append(0)

  if value[3] in urlNG_list:
    ng_url.append(1)
  else:
    ng_url.append(0)

df['pardot'] = use_pardot
df['eloqua'] = use_eloqua
df['marketo'] = use_marketo
df['NGURL'] = ng_url
df.to_csv("./完了①FUMA_東京_小売全般_マスターデータ.csv", encoding="utf_8_sig")
```
